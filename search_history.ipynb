```python
from PyPDF2 import PdfReader, PdfWriter

input_pdf = r'Mobile_dataset.pdf'  # Your 162,000-page PDF
output_pdf = 'search_history.pdf'  # Output subset PDF

start_page, end_page = 28530, 28534  # Example: Extract pages 100 to 200

reader = PdfReader(input_pdf)
writer = PdfWriter()

for i in range(start_page, end_page + 1):
    writer.add_page(reader.pages[i])

with open(output_pdf, "wb") as f:
    writer.write(f)

print("Subset PDF saved as", output_pdf)
```

```python
import pdfplumber
import pandas as pd

# Path to the uploaded PDF
pdf_path = "search_history.pdf"

# Extract text from the PDF
extracted_text = ""
table_data = []

with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            extracted_text += text + "\n"  # Store text
        tables = page.extract_tables()
        if tables:
            table_data.extend(tables)  # Store tables

# Display first 1000 characters of extracted text and first 2 tables if available
extracted_text[:1000], table_data[:2]
```

```python
# Flatten table
flattened_data = [row for table in table_data for row in table]

# Map to PDF column headers
columns = ["#", "Timestamp", "Source", "Value", "Parameters", "Origin", "Service", "Identifier", "Deleted Account"]

# Create DataFrame
df = pd.DataFrame(flattened_data, columns=columns)

# Clean newlines and whitespaces
df = df.applymap(lambda x: x.replace('\n', ' ').strip() if isinstance(x, str) else x)

# Save to CSV
df.to_csv("cleaned_pdf_search_history.csv", index=False)

# Display sample
print(df.head(10))
```

```python
import pandas as pd

# Assume df is already loaded and has the 'Timestamp' column in this format: '15-12-2024 08:17:11(UTC+0)'

# Step 1: Convert 'Timestamp' to datetime format
df['Timestamp'] = df['Timestamp'].str.replace('(UTC+0)', '', regex=False).str.strip()
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d-%m-%Y %H:%M:%S', errors='coerce')

# Step 2: Function to filter data by date range
def filter_by_date(df, start_date_str, end_date_str):
    start_date = pd.to_datetime(start_date_str)
    end_date = pd.to_datetime(end_date_str)
    filtered_df = df[(df['Timestamp'] >= start_date) & (df['Timestamp'] <= end_date)]
    return filtered_df

# Example usage:
start = '2024-10-01 00:00:00'
end = '2024-12-31 23:59:59'

filtered_df = filter_by_date(df, start, end)

# Step 3: Show filtered results
print(filtered_df)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', str(text))  # remove special chars
    tokens = word_tokenize(text.lower())
    return " ".join([word for word in tokens if word not in stop_words])

df['Cleaned_Value'] = df['Value'].apply(preprocess)
```

```python
mental_health_keywords = ['depression', 'suicidal', 'anxiety', 'therapy', 'mental health', 'sad', 'lonely', 'help', 'panic attack']
```

```python
criminal_keywords = ['how to make bomb', 'dark web', 'buy drugs', 'fake id', 'vpn to hide', 'black market', 'escape police']
```

```python
def detect_intent(text, keywords):
    return any(kw in text for kw in keywords)

df['MentalHealth_Flag'] = df['Cleaned_Value'].apply(lambda x: detect_intent(x, mental_health_keywords))
df['CriminalIntent_Flag'] = df['Cleaned_Value'].apply(lambda x: detect_intent(x, criminal_keywords))
```

```python
df['Hour'] = df['Timestamp'].dt.hour
active_hours = df['Hour'].value_counts().sort_index()
```

```python
import subprocess
import pandas as pd

def run_ollama_llm(search_history: str):
    prompt = f"""
You are a forensic analyst AI. Based on this search history, analyze and return:

- Mental health state (if any signs like depression, anxiety, curiosity, etc.)
- Criminal or suspicious behavior (if any)
- Behavioral patterns
- Final summary verdict

Search history:
{search_history}
"""
    result = subprocess.run(
        ['ollama', 'run', 'mistral'],  # or llama3 / zephyr
        input=prompt.encode(),
        stdout=subprocess.PIPE
    )
    return result.stdout.decode()

# Load your dataframe
#df = pd.read_csv(r'C:\Users\Yash lichade\OneDrive\Desktop\final\search_history_cleaned.csv')  # or however your df is set

# Filter to just the 'Value' field (search queries)
search_text = "\n".join(df['Value'].dropna().tolist()[:30])  # Use top 30 for example

# Run the LLM
analysis = run_ollama_llm(search_text)
print(analysis)
```

```python
from langchain_community.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.schema import Document

# Sample docs
docs = [Document(page_content="Person searched for bike rental and RTO challan app.")]

# Load embedding model
embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Build vectorstore
vectorstore = FAISS.from_documents(docs, embedding)

# Perform search
query = "transport related activity"
results = vectorstore.similarity_search(query, k=2)

for res in results:
    print(res.page_content)
```
